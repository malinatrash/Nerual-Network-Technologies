```python
# 1
# Генерация случайных чисел с равномерным распределением
M = 3
uniform_random_numbers = np.random.uniform(0, 10, (M, 2))  # M строк, 2 столбца

# Применение корня квадратного к числам
sqrt_numbers = np.sqrt(uniform_random_numbers)

# Функция Букина (пример)
def bukin_function(x, y):
    return 100 * np.sqrt(np.abs(y - 0.01 * x ** 2)) + 0.01 * np.abs(x + 10)

# Применение функции Букина к числам
bukin_result = bukin_function(uniform_random_numbers[:, 0], uniform_random_numbers[:, 1])

# Создание матрицы аргументов и результатов
arguments_matrix = np.column_stack((uniform_random_numbers, sqrt_numbers))
results_matrix = np.column_stack((uniform_random_numbers, bukin_result))

print("Матрица аргументов (случайные числа и их корни):")
print(arguments_matrix)

print("\nМатрица результатов (аргументы и значения функции Букина):")
print(results_matrix)
```

```python
# 2

from sklearn.model_selection import train_test_split

# Предположим, у вас есть матрица данных "arguments_matrix" и матрица результатов "results_matrix"

# Разделение на обучающую (80%) и тестовую (20%) выборки
X_train, X_test, y_train, y_test = train_test_split(arguments_matrix, results_matrix, test_size=0.2, random_state=42)

# X_train - обучающая выборка для аргументов
# X_test - тестовая выборка для аргументов
# y_train - обучающая выборка для результатов
# y_test - тестовая выборка для результатов

# Для использования этих данных в модели вашего выбора (например, линейной регрессии):
from sklearn.linear_model import LinearRegression

# Создание и обучение модели
model = LinearRegression()
model.fit(X_train, y_train)

# Предсказание на тестовой выборке
predictions = model.predict(X_test)
```

```python
# 3.1 

# Создание модели нейронной сети
model = tf.keras.Sequential([
	# Первый скрытый слой с 10 нейронами
    tf.keras.layers.Dense(10, activation='relu'),
    # Выходной слой с одним нейроном (для аппроксимации одномерной функции)
    tf.keras.layers.Dense(1)
])

# Компиляция модели с выбором оптимизатора и функции потерь
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Оценка модели на тестовых данных
loss = model.evaluate(X_test, y_test)
print("Потери на тестовых данных:", loss)

# Получение предсказаний модели на тестовых данных
predictions = model.predict(X_test)
```

```python
import neurokit2 as nk
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Масштабирование данных
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Создание и обучение RBF-сети
n_centers = 10  # Количество центров (кластеров)
rbf_model = nk.GaussianRBF(n_centers=n_centers)
rbf_model.fit(X_train_scaled, y_train, epochs=50)

# Предсказание на тестовых данных
predictions = rbf_model.predict(X_test_scaled)

# Оценка качества модели (например, среднеквадратичная ошибка)
mse = mean_squared_error(y_test, predictions)
print("Среднеквадратичная ошибка:", mse)
```
